```python
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Folder containing the CSV files
folder_path = r'C:\Users\s5236256\Documents\GitHub\ml4rt\Paper_results\Sample_size_test'

# Initialize an empty list to store dataframes
dataframes = []

# Loop through each file in the folder
for file_name in os.listdir(folder_path):
    if file_name.endswith('.csv'):
        file_path = os.path.join(folder_path, file_name)
        df = pd.read_csv(file_path)
        dataframes.append(df)

# Concatenate all dataframes into a single dataframe
combined_df = pd.concat(dataframes)
```

```python
combined_df
```

```python
# Plotting the 'mean_error' against 'n' as an x-y scatter plot
plt.scatter(combined_df['n'], combined_df['mean_error'])
plt.xlabel('n')
plt.ylabel('mean_error')
plt.title('Mean Error vs. n')
plt.show()
```

```python

# Set font size for better readability
plt.rcParams.update({'font.size': 12})

# Plotting the 'mean_error' against 'n' as an x-y scatter plot
plt.scatter(combined_df['n'], combined_df['mean_error'], color='black', alpha=0.6)

# Add horizontal grid lines
plt.grid(axis='y', which='major', color='gray', linestyle='-', linewidth=0.5)

# Set axis labels
plt.xlabel('Number of training samples')
plt.ylabel('Mean absolute error (m)')

# Adjust plot layout and padding
plt.tight_layout()

# Save the figure as a high-resolution image (optional)
# plt.savefig('scatter_plot.png', dpi=300)

# Display the plot
plt.show()
```
