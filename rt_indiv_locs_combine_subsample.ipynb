```python
import pandas as pd
```

```python
predictions_file = "Example_data\Output\Predictions\MovementEcologyPaper\predictions_active_tags_20231205.xlsx"
indiv_locs_file = r"C:\Users\John\SynologyDrive\JOBS\~2019\QEJ19152\DATA ANALYSIS\MOVEMENT_ECOLOGY\MASTER_DATA\BTF_Individual_Locations_20230718.xlsx"
tag_list_file = r"C:\Users\John\SynologyDrive\JOBS\~2019\QEJ19152\DATA ANALYSIS\MOVEMENT_ECOLOGY\MASTER_DATA\Tag_List.xlsx"

predictions = pd.read_excel(predictions_file)
indiv_locs = pd.read_excel(indiv_locs_file)
tag_list = pd.read_excel(tag_list_file)
```

```python
predictions
```

```python
tag_list
```

```python
predictions = predictions.merge(tag_list[['Report_Tag_ID', 'Band_Number']], left_on='TagID', right_on='Report_Tag_ID', how='left')

```

```python
# Rename some columns
predictions.rename(columns={'latitude_pred': 'POINT_Y', 'longitude_pred': 'POINT_X', 'Band_Number': 'MetalBand'}, inplace=True)
# Add a new column 'Method' with all values set to 'ART'
predictions['Method'] = 'ART'

indiv_locs.rename(columns={'CreationDateAEST': 'DateTime'}, inplace=True)
```

```python
concatenated_df = pd.concat([predictions, indiv_locs], ignore_index=True)
```

```python
concatenated_df.columns
```

```python
# Create a new dataframe with only the specified columns
selected_columns_df = concatenated_df[['MetalBand','DateTime','Report_Tag_ID','Method','BTF_Activity_Report','Veracity','POINT_Y','POINT_X','GUIDCOPY']]
```

```python
# Save the dataframe to a CSV file
selected_columns_df.to_csv('selected_columns_data.csv', index=False)
```
